The engine schematic (your puzzle input) consists of a visual representation of the engine. There are lots of numbers and symbols you don't really understand, but apparently any number adjacent to a symbol, even diagonally, is a "part number" and should be included in your sum. (Periods (.) do not count as a symbol.)

Here is an example engine schematic:

467..114..
...*......
..35..633.
......#...
617*......
.....+.58.
..592.....
......755.
...$.*....
.664.598..
In this schematic, two numbers are not part numbers because they are not adjacent to a symbol: 114 (top right) and 58 (middle right). Every other number is adjacent to a symbol and so is a part number; their sum is 4361.

Of course, the actual engine schematic is much larger. What is the sum of all of the part numbers in the engine schematic?

```python
import numpy as np

def get_adjacent_value(schematic, row, col):
    top=None
    bottom=None
    left=None
    right=None
    left_top=None
    left_bottom=None
    right_top=None
    right_bottom=None
    evaluate_position = []
    
    if row != 0:
        #top
        top = schematic[row-1][col]
        if top.isdigit():
            evaluate_position.append((row-1,col))
    if row != len(schematic)-1:
        #bottom
        bottom = schematic[row+1][col]
        if bottom.isdigit():
            evaluate_position.append((row+1,col))
        
    if col != 0:
        #left
        left = schematic[row][col-1]
        if left.isdigit():
            evaluate_position.append((row,col-1))
        
    if col != len(schematic)-1:
        #right
        right = schematic[row][col+1]
        if right.isdigit():
            evaluate_position.append((row,col+1))
    
    if row != 0 and col != 0:
        #left top
        left_top = schematic[row-1][col-1]
        if left_top.isdigit():
            evaluate_position.append((row-1,col-1))

    if col != 0 and row != len(schematic)-1:
        #left bottom
        left_bottom = schematic[row+1][col-1]
        if left_bottom.isdigit():
            evaluate_position.append((row+1,col-1))

    if col != len(schematic)-1 and row != 0:
        #right top
        right_top = schematic[row-1][col+1]
        if right_top.isdigit():
            evaluate_position.append((row-1,col+1))

    if row != len(schematic)-1 and col != len(schematic)-1:
        #right bottom
        right_bottom = schematic[row+1][col+1]
        if right_bottom.isdigit():
            evaluate_position.append((row+1,col+1))
        
    return evaluate_position


def get_full_digit(schematic,value):
    row, col = value
    index_left = col
    index_right = col
    while schematic[row][index_left-1].isdigit() and index_left!=0:
        index_left-=1
    while index_right!=len(schematic) and schematic[row][index_right].isdigit():
        index_right+=1
    full_digit = schematic[row][index_left:index_right]
    change_index = index_left
    for i in range(index_left,index_right):
        replace_list = list(schematic[row])
        replace_list[change_index]='X'
        schematic[row]=''.join(replace_list)
        change_index+=1

    return schematic, full_digit

unique_char = ''.join(set(input_text))
#to find all unique characters
print(unique_char)
total=0
allowable_char = r'%+@=/&$-#*'

schematic = np.array(input_text.splitlines())
np.savetxt('before_scheme.txt', schematic,fmt='%s', delimiter='')
for i in range(0, len(schematic)):
    for j in range(0, len(schematic[0])):
        if schematic[i][j] in allowable_char:
            adjacent_values = get_adjacent_value(schematic, i, j)
            for value in adjacent_values:
                schematic, full_digit = get_full_digit(schematic, value)
                if full_digit:
                    total+=int(full_digit)

np.savetxt('schematic.txt', schematic,fmt='%s', delimiter='')
print(total)
```
